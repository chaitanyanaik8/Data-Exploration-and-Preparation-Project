# Splitting the PUBG Data(4446966 rows, 29 columns) into Train and Test by 80%-20%

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import drive
import warnings
warnings.filterwarnings("ignore")
```

```python
drive.mount('/content/drive/')
```

```python
Train = pd.read_csv('/content/drive/MyDrive/PUBG-finish-placement-prediction/Original_data/train_V2.csv')
Train.head()
```

```python
X = Train.drop(columns='winPlacePerc')
y = Train['winPlacePerc']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
Train_data = pd.concat([X_train,y_train], axis=1)
```

```python
Test_data = pd.concat([X_test,y_test], axis=1)
```

```python
Train_data.to_csv('/content/drive/MyDrive/PUBG-finish-placement-prediction/Train_data.csv', index=False)
```
